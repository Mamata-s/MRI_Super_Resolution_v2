# Define loss function

```python
loss_fn = nn.MSELoss()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# hyperparameters
batch_size = 32
channels = 1
n_epochs = 10
learning_rate = 0.01

# lbda = 0.3
# alpha = 0.5
# beta = 0.2
```

# Training a simple network

```python
model = mconcat.CoarseSRNetwork()
model = nn.DataParallel(model,device_ids=[0,1])
model.to(device)


# tb = SummaryWriter()

# images, labels,parsers = next(iter(dataloaders))
# grid = torchvision.utils.make_grid(images)
# grid_labels = torchvision.utils.make_grid(labels)
# tb.add_image("images", grid)
# tb.add_image("labels", grid_labels)

# model.eval()
# with torch.no_grad():
#     tb.add_graph(model,images)

optim = torch.optim.SGD(params = model.parameters(), lr = learning_rate, momentum=0.9)

model.train() 
losses = []
k = 0
for epoch in range(1, n_epochs + 1):
    for idx, (images, labels,parsers) in enumerate(dataloaders):
        images = images.to(device)
        labels = labels.to(device)
        parsers = parsers.to(device)
        torch.cuda.empty_cache()

        outputs = model(images)

        loss = loss_fn(outputs, labels) 

        optim.zero_grad()
        loss.backward()
        optim.step() 

        losses.append(float(loss))
#         tb.add_scalar("Loss", float(loss), idx)
        
        if idx % 200 == 0:
            print("Epoch [%d/%d]. Iter [%d/%d]. Loss: %0.2f" % (epoch, n_epochs, idx + 1, len(dataloaders), loss))
#     for name, weight in model.named_parameters():
#         tb.add_histogram(name, weight, epoch)
#         tb.add_histogram(f'{name}.grad',weight.grad, epoch)

    if k == 25:
        path = f'model/simple_model/model_full_'+str(epoch)+'.pth'
        torch.save(model.state_dict(), path)
        k=0
    k+=1
# tb.close()
```

# Evaluate model

```python
input_image_dir = 'outputs/simple_model/inputs'
output_image_dir = 'outputs/simple_model/outputs'

l1_error =[]
l2_error = []
psnr =[]
rmse = []
ssim=[]

dict_eval = ut.create_dictionary(input_image_dir,output_image_dir)

for keys in dict_eval.keys():
    input_image = np.array(Image.open(input_image_dir+'/'+keys).convert('L'))
    output_image = np.array(Image.open(output_image_dir+'/'+dict_eval[keys]).convert('L'))
    
    input_image = ut.min_max_normalize(input_image)
    output_image = ut.min_max_normalize(output_image)
    
    l1 = err.calculate_l1_distance(input_image,output_image)
    l2 = err.calculate_l2_distance(input_image,output_image)
    psnr_val = err.calculate_RMSE(input_image,output_image)
    rmse_val = err.calculate_PSNR(input_image,output_image)
    (ssim_score, diff) = compare_ssim(input_image, output_image, full=True)
    diff = (diff * 255).astype("uint8")
    
    l1_error.append(l1)
    l2_error.append(l2)
    psnr.append(psnr_val)
    rmse.append(rmse_val)
    ssim.append(ssim_score)
```

```python
input_image_dir = 'outputs/simple_model/full_img_model/inputs'
output_image_dir = 'outputs/simple_model/full_img_model/outputs'

l1_error_full =[]
l2_error_full = []
psnr_full =[]
rmse_full = []
ssim_full = []

dict_eval_full = ut.create_dictionary(input_image_dir,output_image_dir)

for keys in dict_eval_full.keys():
    input_image = np.array(Image.open(input_image_dir+'/'+keys).convert('L'))
    output_image = np.array(Image.open(output_image_dir+'/'+dict_eval_full[keys]).convert('L'))
    
    input_image = ut.min_max_normalize(input_image)
    output_image = ut.min_max_normalize(output_image)
    
    l1 = err.calculate_l1_distance(input_image,output_image)
    l2 = err.calculate_l2_distance(input_image,output_image)
    psnr_val = err.calculate_RMSE(input_image,output_image)
    rmse_val = err.calculate_PSNR(input_image,output_image)
    (ssim_score, diff) = compare_ssim(input_image, output_image, full=True)
    diff = (diff * 255).astype("uint8")
    
    l1_error_full.append(l1)
    l2_error_full.append(l2)
    psnr_full.append(psnr_val)
    rmse_full.append(rmse_val)
    ssim_full.append(ssim_score)
```

```python
fig = plt.figure(figsize=(13,25))

mm = [l1_error,l2_error,psnr,rmse,ssim]
full = [l1_error_full,l2_error_full,psnr_full,rmse_full,ssim_full]
k=0

names = [' L1 error','L2 error','RMSE','PSNR','SSIM']
for idx in np.arange(0,10,2):
    ax = fig.add_subplot(10/2, 2, idx+1)
    ax.plot(mm[k])
    ax.title.set_text('model : '+ str(names[k]))
    
    ax = fig.add_subplot(10/2,2, idx+2)
    ax.plot(full[k])
    ax.title.set_text('model with full img input : '+ str(names[k]))
    k+=1
```
