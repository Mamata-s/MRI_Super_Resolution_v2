```python
import numpy as np
# from pylab
# import *
import nibabel as nib
import torch
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import warnings
warnings.filterwarnings("ignore")

def load_data_nii(fname):
    img = nib.load(fname)
    affine_mat=img.affine
    hdr=img.header
    data = img.get_fdata()
    data_norm = data
    data_norm = torch.from_numpy(data_norm)
    return data_norm 
```

# Matlab style

```python
# define gaussian filter
def matlab_style_gauss2D(shape=(3,3),sigma=0.5):
    """
    2D gaussian mask - should give the same result as MATLAB's
    fspecial('gaussian',[shape],[sigma])
    """
    m,n = [(ss-1.)/2. for ss in shape]
    y,x = np.ogrid[-m:m+1,-n:n+1]
    h = np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
    h[ h < np.finfo(h.dtype).eps*h.max() ] = 0
    sumh = h.sum()
    if sumh != 0:
        h /= sumh
    return h
```

# using PIL Gaussian filter

```python

# ImageFilter for using filter() function
from PIL import Image, ImageFilter
  
# Opening the image 
# (R prefixed to string in order to deal with '\' in paths)
image = Image.open(r"OneDrive/AD_P522R_F1_160/hr_cropped_images/cropped_hr24.png")
  
# Blurring image by sending the ImageFilter.
# GaussianBlur predefined kernel argument
image = image.filter(ImageFilter.GaussianBlur)
  
# Displaying the image
image.show()
```

```python
def gauss(x,y, sigma):

    return 1 / (2 * np.pi * sigma**2) * np.exp(-(x**2+ y**2) / (2 * sigma**2))
```

```python
H = np.zeros((720,512))
sigma = 5
H.shape
for i in range(H.shape[0]):
    for j in range(H.shape[1]):
        H[i][j] = gauss(i,j,sigma)
```

```python
G=matlab_style_gauss2D((720,512),sigma)
```

```python
import matplotlib.pyplot as plt

figure(figsize=(8, 6), dpi=80)
plt.subplot(221),plt.imshow(H,cmap = 'gray')
plt.title('custom gauss function filter'),plt.xticks([]), plt.yticks([])

plt.subplot(222),plt.imshow(G,cmap = 'gray')
plt.title('matlab style gauss'),plt.xticks([]), plt.yticks([])
plt.show()
```

```python
FH = np.fft.fft2(H)
fhshift = np.fft.fftshift(FH)
fhmagnitude_spectrum = 20*np.log(np.abs(fhshift))
fhmagnitude__without_spectrum = 20*np.log(np.abs(FH))

FG = np.fft.fft2(G)
fgshift = np.fft.fftshift(FG)
fgmagnitude_spectrum = 20*np.log(np.abs(fgshift))
fgmagnitude__without_spectrum = 20*np.log(np.abs(FG))

plt.figure(figsize=(19,12))
plt.subplot(221),plt.imshow(H, cmap = 'gray')
plt.title('Input Custom Filter'), plt.xticks([]), plt.yticks([])

plt.subplot(222),plt.imshow(fhmagnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum of Custom Filter'), plt.xticks([]), plt.yticks([])


plt.subplot(223),plt.imshow(G, cmap = 'gray')
plt.title('Input matlab style Filter'), plt.xticks([]), plt.yticks([])

plt.subplot(224),plt.imshow(fgmagnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum of matlab style Filter'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
F=np.fft.fft2(data_hr[:,:,80])

FH = np.fft.fft2(H)
FFH = F*FH
ffhshift = np.fft.fftshift(FFH)
img_recon_h = np.abs(np.fft.ifft2(np.fft.ifftshift(ffhshift)))

FG = np.fft.fft2(G)
FFG = F*FG
ffgshift = np.fft.fftshift(FFG)
img_recon_g = np.abs(np.fft.ifft2(np.fft.ifftshift(ffgshift)))

plt.figure(figsize=(25,19))
plt.subplot(221),plt.imshow(img_recon_h, cmap = 'gray')
plt.title('Convolution with custom filter'), plt.xticks([]), plt.yticks([])


plt.subplot(222),plt.imshow(img_recon_g, cmap = 'gray')
plt.title('Convolution with matlab style filter'), plt.xticks([]), plt.yticks([])

plt.show()
```
